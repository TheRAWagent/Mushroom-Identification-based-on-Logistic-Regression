```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import roc_curve,auc,confusion_matrix

%matplotlib inline
```

```python
data=pd.read_csv('mushrooms.csv')
data.head()
```

```python
x=data['class']
ax=sns.countplot(x=x, data=data)
```

```python
def plot_data(hue,data):
    for i,col in enumerate(data.columns):
        plt.figure(i)
        ax=sns.countplot(x=data[col],hue=hue,data=data)
```

```python
for col in data.columns:
    print(f"{col} : {data[col].isnull().sum()}")
```

```python
le = LabelEncoder()
data['class']=le.fit_transform(data['class'])
data.head()
```

```python
encoded_data=pd.get_dummies(data)
encoded_data.head()
```

```python
Y=data['class'].values.reshape(-1,1)
X=encoded_data.drop(['class'],axis=1)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_reg = LogisticRegression()
logistic_reg.fit(X_train,Y_train.ravel())
y_prob = logistic_reg.predict_proba(X_test)[:,1]
y_pred=np.where(y_prob>0.5,1,0)
```

```python
log_confusion_matrix = confusion_matrix(Y_test,y_pred)
print(log_confusion_matrix)
```

```python
falsePositivityRate,truePositivityRate,threshold=roc_curve(Y_test,y_prob)
roc_auc=auc(falsePositivityRate,truePositivityRate)
print(f"{roc_auc}")
```

```python
def plot_roc(roc_auc):
    plt.figure(figsize=(7,7))
    plt.plot(falsePositivityRate,truePositivityRate,color='red',label='AUC = %0.2f' % roc_auc)
    plt.plot([0,1],[0,1],linestyle='--')
    plt.axis('tight')
plot_roc(roc_auc)
```
